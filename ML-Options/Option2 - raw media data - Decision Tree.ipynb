```python
c = medianeconomics_df.Genre.astype('category')

d = dict(enumerate(c.cat.categories))
print (d)
```

```python
#Label encoding 

medianeconomics_df['Genre'] =medianeconomics_df['Genre'].astype('category').cat.codes
```

```python
# Defining the features set.
X = medianeconomics_df.copy()
X = X.drop("Genre", axis=1)
X.head()
```

```python
# Define the target set.
y = medianeconomics_df["Genre"].values
y[:5]
```

```python
#Split the Data Into Training and Testing Sets

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
#Scale of the training and testing data

# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Defining model

# Creating the decision tree classifier instance.
model = tree.DecisionTreeClassifier()
# Fitting the model.
model = model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = model.predict(X_test_scaled)
predictions
```

```python
#ploting the decision tree

tree.plot_tree(model);
```

```python
#decision tree more interpretable 

fn=['inflation','SPyield','unemployment','UMCSENT', 'homeindex', 'Gross']
cn=['Action', 'Adventure', 'Comedy', 'Drama', 'Horror', 'Thriller/Suspense']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

```python
#Evaluating the model 

# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm)
cm_df
```
